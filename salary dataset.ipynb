# mean of salary according to their job title

```python
g=df.groupby('Job Title').agg({"Salary":'mean'})
print(g)
```

# Heatmap of the mean of salary according to their job title

```python
sns.heatmap(g)
plt.title("Mean Salary-Job Title distribution",c="b")
```

# counting the no.of employees in a country 

```python
groupA=df.loc[(df['Country']=='USA')].count()
groupB=df.loc[(df['Country']=='UK')].count()
groupC=df.loc[(df['Country']=='Canada')].count()
groupD=df.loc[(df['Country']=='China')].count()
groupE=df.loc[(df['Country']=='Australia')].count()

print(groupA)
```

# pie chart Distribution of Contribution employees in a Country

```python
l=["groupA","groupB","groupC","groupD","groupE"]
mlist=[groupA["Country"],groupB["Country"],groupC["Country"],groupD["Country"],groupE["Country"]]

print(mlist)
plt.pie(mlist,labels=l,autopct="%1.2f%%")
plt.title("Distribution of Contribution employees in a Country")
plt.show()
```
